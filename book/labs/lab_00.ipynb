### 🐍 A quick look with Python (optional, 🔧)

Later in the semester you'll build maps with Python directly — this is just a preview. The cell below draws a simple map centered on Silver City using `leafmap`, an open-source Python mapping library we'll use again later in the course.

```python
# 🔧 OPTIONAL — a first taste of Python-based mapping (we'll do much more with this later in the semester)
# If leafmap isn't installed, uncomment the line below:
# !pip install leafmap -q

import leafmap

silver_city = (32.7701, -108.2803)  # (latitude, longitude)

m = leafmap.Map(center=silver_city, zoom=11)
m.add_basemap("Esri.WorldImagery")
m
```
